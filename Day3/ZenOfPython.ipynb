#### Some good practices:
1. Clean up your code before you submit it.
    - Remove unnecessairy cells or lines of code you really don't need
    - Make sure only necessary things remain
2. Comment your code.
    - At the top it is nice to have a heading that contains the code name, author, and maybe some information about what the code does.
    - Within the code, add comments to parts of the code that might be confusing or hard to understand.
    - Don't go totally crazy with comments, you can comment too much.
3. Use descriptive variables names
    - The best code is readable and easy to understand.
    - The more your variables describe what is contained, the better.
