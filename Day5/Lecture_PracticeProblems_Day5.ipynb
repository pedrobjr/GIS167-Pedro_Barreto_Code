#### If statements 

These are super important to writing code. So often you want to only do something in your code IF something is true. 

```python
# Lets update the guess my number game:

# Guess the computers number
# Guess a number between 1-10 (integers)
# This time let the computer choose the number
import numpy as np
computers_number = np.random.randint(11) #generate a random number between 0 and 10
your_guess = 10

# This time we can say you won!
print(f"Your guess was {your_guess}," \
      f" the computers number was {computers_number}.")

if computers_number == your_guess:
    print("You Win!")
```

```python
# Code that uses a FOR loop and an IF statement
user_names = ['Joanna','Kristin','Ross','Stefanie','David','Bongo']

for name in user_names:
    print('-------------------------')
    print(f"Hello there {name}")
    if name == "Bongo":
        print("You are a cat!")
```

#### If-Else statements 

These are great for doing different things depending on an if statement. 

```python
# Lets update the guess my number game:

# Guess the computers number
# Guess a number between 1-10 (integers)
# This time let the computer choose the number
import numpy as np
computers_number = np.random.randint(11) #generate a random number between 0 and 10
your_guess = 10

# This time we can say you won!
print(f"You guess was {your_guess}," \
      f" the computers number was {computers_number}.")

if computers_number == your_guess:
    print("You Win!")
else:
    print('You Lose!')
```

```python
# Update the for loop:

user_names = ['Joanna','Kristin','Ross','Stefanie','David','Bongo']

for name in user_names:
    print('-------------------------')
    print(f"Hello there {name}")
    if name == "Bongo":
        print("You are a cat!")
    else:
        print("You are a human!")
```
