```python
# Conditional Checks
# Equal ==
# Not Equal !=
# Less Than <
# Greater Than >
# Less Than or Equal <=
# Greater Than or Equal >=
# AND --> All things are True
# OR --> At least one of the things are True

dog = "Stormy"
print('This should be True')
print(dog == "Stormy")
age = 19
print('This should be False')
print(age <= 19)
```

```python
colors = ['red','green','BLUE','Purple','black']
print('red' in colors)
print('pink' in colors)
print('BLUE'.lower() in colors)
print('red' in colors and 'great' in colors)
print('red' in colors or 'pink' in colors)
```

```python
door_number = 2
prizes = ['car','vacation to Cancun','dirty mop']
for prize in {range(len(prizes))}:
    if door_number == 1:
        print(f'You just won a {prizes[0]}!')
    if door_number == 2:
        print(f'You just won a {prizes[1]}!')
    if door_number == 3:
        print(f'You just won a {prizes[2]}!')
```

```python
my_things = ['cup','sandwich','computer','hotdog','watch']
can_i_eat_it = ['not food','food','not food','food','not food']
for i in range(len(my_things)):
    if can_i_eat_it[i] == 'food':
        print(f'I like to eat my {my_things[i]}')
    if can_i_eat_it[i] == 'not food':
        print(f'I can use a {my_things[i]}')
```
